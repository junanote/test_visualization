남한과 북한 전력 사용량 비교

화력 발전량 비교 : 2000-2016

수력 발전량 비교 : 1990-2014

둘다 그래프작성 : 종류선택 이탤릭체 텍스트


```python
import pandas as pd
df= pd.read_excel('남북한발전전력량.xlsx')
df
```

```python
df_w = df.loc[[1,6],'2000':'2016']
df_w.index =['South','North']
df_w
```

```python
df_w = df_w.transpose()
df_w.plot(kind='line')
```

```python
df_f = df.loc[[2,7],'1990':'2014']
df_f.index =['South','North']
df_f
```

```python
df_f = df_f.transpose()
df_f.plot(kind='bar')
```
